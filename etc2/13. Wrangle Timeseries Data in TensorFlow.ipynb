이제 데이터를 출력하는 함수를 만들어 보자. 디펄트로 데이터셋의 처음 5개만 출력하는...

```python
def printDs(ds, take=5):
    for ex in ds.take(take):
        print(ex)
```

간단한 tf dataset을 만들어보자.

```python
dataset = tf.data.Dataset.from_tensor_slices(seq_expanded)
```

```python
dataset
```

```python
printDs(dataset)
```

```python
printDs(dataset,40)
```

### Dataset window

```python
inputCount = 5
outputCount = 1
windowSize = inputCount + outputCount
```

```python
dataset = dataset.window(windowSize,shift=1,drop_remainder=True) # 윈도우 메서드는  데이터셋을 윈도우 리스트로 만듬
```

```python
dataset
```

window 메서드를 통해 데이터가 WindowDataset 타입으로 바꼈다. 다음 두가지에 유의하자.    
- 마지막 6개는 사용되지 않는다.   
- 데이터는 윈도우로 되어 있어서 윈도우들을 배치형태로 만들어줘야 한다.

```python
printDs(dataset)
```

배치 만들기

```python
dataset = dataset.flat_map(lambda b: b.batch(windowSize))
printDs(dataset)
```

이제 각 윈도우가 배치로 되었다.  
이제 입력과 출력을 나누다. 각 윈도우에 대하여 map 함수를 사용한다.   
마지막 element가 output이고 처음 5개 element가 입력이다.

```python
dataset = dataset.map(lambda x: (x[:-1],x[-1]))
printDs(dataset)
```

시퀀스를 셔플한다. 학습 성능 개선을 위해서

```python
dataset = dataset.shuffle(100)
printDs(dataset)
```

이제 학습을 위한 배치를 나눠보자.

```python
dataset = dataset.batch(3)
printDs(dataset)
```
