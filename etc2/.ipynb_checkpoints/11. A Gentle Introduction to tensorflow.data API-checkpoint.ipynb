그러나 두 컬럼으로 wind vector를 만들면 두 

```python
wv = df.pop('wv (m/s)')
max_wv = df.pop('max. wv (m/s)')

# Convert to radians.
wd_rad = df.pop('wd (deg)')*np.pi / 180

# Calculate the wind x and y components.
df['Wx'] = wv*np.cos(wd_rad)
df['Wy'] = wv*np.sin(wd_rad)

# Calculate the max wind x and y components.
df['max Wx'] = max_wv*np.cos(wd_rad)
df['max Wy'] = max_wv*np.sin(wd_rad)
```

### Data windowing

본 튜토리얼에서 모델은 연속 샘플의 윈도우를 기반으로 예측을 진행한다.   

본 섹션에서는 여러 모델에서 windowing 방법이 재사용 될수 있도록 포커스를 맞춘다.  

1. 예를 들어 24시간의 기록이 주어지고 미래의 24시간 뒤의 단일 예측을 하려면 다음과 같이 정의할 수 있다.

2. 6시간의 입력이 주어지고 바로 다음 스텝의 1시간을 예측하는 모델은 다음과 같은 윈도우가 필요하다.

여기서는 WindowsGenerator라는 클래스를 정의한다. 이 것은 다음과 같은 기능이 있다.   
- 위 다이어 그램에서 보여준 인덱스와 오프셋을 다룬다.   
- 피처의 윈도우를 (features, labels) 짝으로 나눈다.  
- 남아있는 윈도우들의 내용를 시각화 할 수 있다.  
- tf.data.Dataset을 사용해서 효율적으로 윈도우 배치를 생성한다.

#### 1. Indexes and offsets  

```python
class WindowGenerator():
    def __init__(self, input_width, label_width, shift,
               train_df=train_df, val_df=val_df, test_df=test_df,
               label_columns=None):
        # Store the raw data.
        self.train_df = train_df
        self.val_df = val_df
        self.test_df = test_df
        
        # Work out the label column indices.
        self.label_columns = label_columns
        if label_columns is not None:
            self.label_columns_indices = {name: i for i, name in enumerate(label_columns)}
        self.column_indices = {name: i for i, name in enumerate(train_df.columns)}
        
        # Work out the window parameters.
        self.input_width = input_width
        self.label_width = label_width
        self.shift = shift
        
        self.total_window_size = input_width + shift

        self.input_slice = slice(0, input_width)
        self.input_indices = np.arange(self.total_window_size)[self.input_slice]

        self.label_start = self.total_window_size - self.label_width
        self.labels_slice = slice(self.label_start, None)
        self.label_indices = np.arange(self.total_window_size)[self.labels_slice]
        
    def __repr__(self):
        return '\n'.join([
                f'Total window size: {self.total_window_size}',
                f'Input indices: {self.input_indices}',
                f'Label indices: {self.label_indices}',
                f'Label column name(s): {self.label_columns}'])
```

#### 2. Split   
연속 입력이 주어졌을 때, split_window 함수는 이 것을 입력 윈도우와 입력 레이블로 만든다.

예를 들어 w2는 아래와 같다.

- 위 그림은 데이터의 피처는 표시하지 않았지만, split_window 함수는 label 컬럼도 처리하므로 단일 출력뿐만 아니라 다중 출력 예제에서도 사용할수 있다.

```python
def split_window(self, features):
    inputs = features[:, self.input_slice, :]
    labels = features[:, self.labels_slice, :]
    if self.label_columns is not None:
        labels = tf.stack(
        [labels[:, :, self.column_indices[name]] for name in self.label_columns],
        axis=-1)

    # Slicing doesn't preserve static shape information, so set the shapes
    # manually. This way the `tf.data.Datasets` are easier to inspect.
    inputs.set_shape([None, self.input_width, None])
    labels.set_shape([None, self.label_width, None])

    return inputs, labels

WindowGenerator.split_window = split_window
```

```python
def plot(self, model=None, plot_col='T (degC)', max_subplots=3):
    inputs, labels = self.example
    plt.figure(figsize = (12, 8))
    plot_col_index = self.column_indices[plot_col]
    max_n = min(max_subplots, len(inputs))
    
    for n in range(max_n):
        plt.subplot(max_n, 1, n+1)
        plt.ylabel(f'{plot_col} [normed]')
        plt.plot(self.input_indices, inputs[n, :, plot_col_index],
                 label='Inputs', marker='.', zorder=-10)
    
        if self.label_columns:
            label_col_index = self.label_columns_indices.get(plot_col, None)
        else:
            label_col_index = plot_col_index
            
        if label_col_index is None:
            continue

        plt.scatter(self.label_indices, labels[n, :, label_col_index],
                edgecolors='k', label='Labels', c='#2ca02c', s=64)
        if model is not None:
            predictions = model(inputs)
            plt.scatter(self.label_indices, predictions[n, :, label_col_index],
                  marker='X', edgecolors='k', label='Predictions',
                  c='#ff7f0e', s=64)
        
        if n == 0:
            plt.legend()
            
    plt.xlabel('Time [h]')
    
WindowGenerator.plot = plot
```

#### 4. Create tf.data.Datasets

마지막으로 make_dataset 메서드로 시계열 데이터 프레임을 (input_window, label_window) 형태의 tf.data.Dataset으로 바꾼다.  
이 때 tf.keras.utils.timeseries_dataset_from_array 함수를 사용한다.

```python
def make_dataset(self, data):
    data = np.array(data, dtype=np.float32)
    ds = tf.keras.utils.timeseries_dataset_from_array(
      data=data,
      targets=None,
      sequence_length=self.total_window_size,
      sequence_stride=1,
      shuffle=True,
      batch_size=32,)

    ds = ds.map(self.split_window) # 윈도우 나누기, 앞에서 만든 split_window 함수 사용해서

    return ds

WindowGenerator.make_dataset = make_dataset
```

- 결국 다른 것 안쓰고 이 함수 사용...

결국 WindowGenerator 객체는 학습, 검증, 시험용 데이터를 가지게 된다.   

앞에서 정의한 make_dataset을 사용하여 tf.data.Datasets로 액세스하기 위한 속성을 추기한다.  
또한 쉽게 액세스하고 플로팅할 수 있도록 example 배치를 추가한다.

```python
@property
def train(self):
    return self.make_dataset(self.train_df)

@property
def val(self):
    return self.make_dataset(self.val_df)

@property
def test(self):
    return self.make_dataset(self.test_df)

@property
def example(self):
    """Get and cache an example batch of `inputs, labels` for plotting."""
    result = getattr(self, '_example', None)
    if result is None:
    # No example batch was found, so get one from the `.train` dataset
        result = next(iter(self.train))
    # And cache it for next time
        self._example = result
    return result

WindowGenerator.train = train
WindowGenerator.val = val
WindowGenerator.test = test
WindowGenerator.example = example
```

이제 WIndowGenerator가 tf.data.Dataset 객체에 접근 가능하다.

### Baseline

가장 간단한 모델을 먼저 만들어 보는 것이 좋다.  
그것은 현재 온도를 포함한 피처로 한시간 뒤 온도를 예측하는 것이다.  

온도 변화는 느리므로 한시간 뒤 온도가 현재와 같다는 모델을 만들어 보자.
당연히 모델은 잘 작동하지 않는다.

```python
class Baseline(keras.Model):
    def __init__(self, label_index = None):
        super().__init__()
        self.label_index = label_index
        
    def call(self, inputs):
        if self.label_index is None:
            return inputs
        result = inputs[:, :, self.label_index]
        return result[:, :, tf.newaxis]
```

이 모델을 인스턴스화 하고, 평가해보자.

### Linear model

학습 기능이 들어갈만한 가장 간단한 모델은 선형 변환 기능을 넣는것이다.(입력과 출력 사이에)

activation이 없는 keras.layers.Dense 레이어 셋은 선형모델이다. 레이어는 (batch, time, inputs) 데이어의 마지막축?을 변환해서 (batch, time, units)로 만든다?    

A tf.keras.layers.Dense layer with no activation set is a linear model. The layer only transforms the last axis of the data from (batch, time, inputs) to (batch, time, units); it is applied independently to every item across the batch and time axes.

```python
linear = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1)
])
```

```python
MAX_EPOCHS = 20
```

```python
def compile_and_fit(model, window, patience=2):
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                                    patience=patience,
                                                    mode='min')

    model.compile(loss=tf.keras.losses.MeanSquaredError(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=[tf.keras.metrics.MeanAbsoluteError()])

    history = model.fit(window.train, epochs=MAX_EPOCHS,
                      validation_data=window.val,
                      callbacks=[early_stopping])
    return history
```

```python
history = compile_and_fit(linear, single_step_window)

val_performance['Linear'] = linear.evaluate(single_step_window.val)
performance['Linear'] = linear.evaluate(single_step_window.test, verbose=0)
```

### RNN

RNN은 시계열을 단계적으로 처리한다. 다시말해 time step에서 time step으로 내부 상태를 유지한다.

여기서는 LSTM을 사용한다.   
LSTM과 같은 RNN 계열의 레이어에서 중요한 생성자 argument는 return_sequences이다.

1) False로 세팅하면(default), 레이어는 마지막 스텝의 출력만 리턴한다. 

2) True로 세팅하면 각 입력에 대한 출력을 내보낸다.   
이 설정은 Stacking RNN이나, multiple time steps 예측모델을 학습할 때 유용하다.

```python
lstm_model = keras.models.Sequential([
    # Shape [batch, time, features] => [batch, time, lstm_units]
    tf.keras.layers.LSTM(32, return_sequences=True),
    # Shape => [batch, time, features]
    tf.keras.layers.Dense(units=1)
])
```

- return_sequences를 True로 설정하면 24시간 분량의 데이터에 대해 모델을 훈련할 수 있다.

그런데 첫단계에서 모델은 이전 단계의 정보를 모르므로 앞의 선형모델이나 dense모델보다 성능이 좋지 못할 수도있다.

```python
print('Input shape:', wide_window.example[0].shape)
print('Output shape:', lstm_model(wide_window.example[0]).shape)
```

Advanced: Residual connections

Baseline 모델은 시계열이 급격하게 변화하지 않는다는 가설에 근거해서 동작했고, 실제로 잘 맞았다.  
여기에 모델들은 랜덤으로 초기화되었고, 출력이 이전 스텝에서 조금 변하는 것으로부터 학습을 했다.   

초기화를 잘 하는 방법도 있지만, 모델 구조를 개선해서 반영하는 방법도 있다.  
시계열 분석에서 다음 값을 예측하기 보다는 다음 시간 단계에서 값이 어떻게 변할지 예측하는 모델을 구축하는 것이 일반적이다.  
마찬가지로 ResNet은 각 레이어가 모델의 누적 결과에 추가되는 아키텍처?를 나타낸다.   

이를 통해 모델이 값의 변화가 작다는 것을 학습하도록 한다.

기본적으로 이 것은 baseline과 일치하도록 초기화 한다. 이 를통해 모델이 빨리 수렴하고, 높은 성능을 내도록 돕는다.   
이 접근 법은 앞에 모든 모델에 사용이 가능하다.  
LSTM에도 사용가능하다.  

그리고 tf.initializers.zeros를 사용하여 초기의 예측 변경(predicted changes)가 작게하고, residual connection을 압도하지 않도록 한다.  
이 구조는 마지막 레이어에만 사용되므로 gradient에 대해 대칭을 깨는 문제가 없다.

```python
class ResidualWrapper(keras.Model):
    def __init__(self, model):
        super().__init__()
        self.model = model

    def call(self, inputs, *args, **kwargs):
        delta = self.model(inputs, *args, **kwargs)

    # The prediction for each time step is the input
    # from the previous time step plus the delta
    # calculated by the model.
        return inputs + delta
```

```python
%%time
residual_lstm = ResidualWrapper(
    keras.Sequential([
    keras.layers.LSTM(32, return_sequences=True),
    keras.layers.Dense(
        num_features,
        # The predicted deltas should start small.
        # Therefore, initialize the output layer with zeros.
        kernel_initializer=tf.keras.initializers.zeros())
]))

history = compile_and_fit(residual_lstm, wide_window)

val_performance['Residual LSTM'] = residual_lstm.evaluate(wide_window.val)
performance['Residual LSTM'] = residual_lstm.evaluate(wide_window.test, verbose=0)
print()
```
